```python
# Setting up the incremental step range for saving scrape to CSV
m = range(0, 4846, 50)
M = range(50, 4846, 50)
steps = zip(m, M)
steps.append((4800, 4816))#get the last bit
steps = steps[-1:]
```

```python
#Scrape for user reviews while saving to CSV...
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

user_reviews = []

#For starting and ending point in tuple of steps
for s,e in steps:

    #Scan through each range of 'steps'
    for i in range(s, e):
        url = 'http://www.metacritic.com/game/pc/'
        #Get the url based on the range of steps
        response = requests.get(url + str(gamesdf['url'][i]) + '/user-reviews', headers=headers)
        soup = BeautifulSoup(response.content, 'lxml')

        #wait
        time.sleep(1)

#         if response.status_code == 404:
#             continue
#         if soup.find(('ol'), {'class':'reviews user_reviews'}) == None:
#             continue

        #Get the number of review pages to comb through... If none go ahead
        if soup.find(('li'), {'class':'page last_page'}) == None:
            pages = 'none'
        else:
            pages = int(soup.find(('li'), {'class':'page last_page'}).find('a').text)

        if pages == 'none':
            #Show progress
            print('Scanning Page: ' + str(i) + ' of ' + str(len(gamesdf))+ ' | ' + str(gamesdf['url'][i]))
            
            #continue on if no reviews are present
            if soup.find(('ol'), {'class':'reviews user_reviews'}) == None:
                continue
            else:
                #Scrape information from the page and place it into a dictionary 'reviews'
                for row in soup.find(('ol'), {'class':'reviews user_reviews'}).find_all(('div'), {'class':'review_stats'}):
                    reviews = {}
                    #find the userId
                    reviews['userId'] = row.find(('div'), {'class':'name'}).text.strip()
                    #find the userScore
                    reviews['userScore'] = row.find(('div'), {'class':'review_grade'}).text.strip()
                    #attach url as id
                    reviews['url'] = str(gamesdf['url'][i])
                    user_reviews.append(reviews)
        else:
            for subpage in range(1, pages):
                url = 'http://www.metacritic.com/game/pc/'
                response = requests.get(url + str(gamesdf['url'][i]) + '/user-reviews?page=' + str(subpage), headers=headers)
                soup = BeautifulSoup(response.content, 'lxml')

                #wait to load...
                time.sleep(1)
                
                #continue on if no reviews are present
                if soup.find(('ol'), {'class':'reviews user_reviews'}) == None:
                    continue
                else:
                    print('Scanning Page: ' + str(i) + ' SubP: ' + str(subpage) + ' of ' + str(len(gamesdf))+ ' | ' + str(gamesdf['url'][i]))

                    #Scrape information from the subpages and place it into a dictionary 'reviews'
                    for row in soup.find(('ol'), {'class':'reviews user_reviews'}).find_all(('div'), {'class':'review_stats'}):
                        reviews = {}
                        #find the userId
                        reviews['userId'] = row.find(('div'), {'class':'name'}).text.strip()
                        #find the userScore
                        reviews['userScore'] = row.find(('div'), {'class':'review_grade'}).text.strip()
                        #attach url as id
                        reviews['url'] = str(gamesdf['url'][i])
                        user_reviews.append(reviews)
                    
                    
    userList = pd.DataFrame(user_reviews)
    userList.to_csv('userReviews5.csv', index=False, encoding='utf-8')
```
