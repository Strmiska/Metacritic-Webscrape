```python
# Setting up the incremental step range for saving scrape to CSV
m = range(0, 4846, 50)
M = range(50, 4846, 50)
steps = zip(m, M)
steps.append((4800, 4816))#get the last bit

steps = steps[38:]
```

```python
gameinfo = []


for s,e in steps:
    
    for i in range(s, e):
        url = 'http://www.metacritic.com/game/pc/' + str(games_df['url'][i])
        headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'lxml')

        print('Scanning page: ' + str(i) + ' of ' + str(len(games_df))+ '|' + str(games_df['url'][i]))

        info = {}
        
        
        if response.status_code != 200:
            continue
        else:
            if str(soup.find(('li'), {'class':'summary_detail product_genre'}).find(('span'), {'class':'data'})) == 'None':
                info['genre'] = 'NaN'
                info['url'] = str(games_df['url'][i])
            else:
                info['genre'] = str(soup.find(('li'), {'class':'summary_detail product_genre'}).find(('span'), {'class':'data'}).text)
                info['url'] = str(games_df['url'][i])

            gameinfo.append(info)
    
    infoTable = pd.DataFrame(gameinfo)
    infoTable.to_csv('gameInfo2.csv', index=False, encoding='utf-8')
```
